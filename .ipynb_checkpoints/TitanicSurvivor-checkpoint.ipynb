```python
# This notebook is the Kaggle Titanc problem: https://www.kaggle.com/c/titanic/data
import os
import pandas as pd

DATASET_PATH = "datasets/Titanic"
DATASET_NAME = "test.csv"
DATASET_URL = DATASET_PATH + DATASET_NAME

def fetch_data(dataset_url=DATASET_URL, dataset_path=DATASET_PATH):
    if not os.path.isdir(dataset_path):
        os.makedirs(dataset_path)

test_dataset = fetch_data()

def load_data(dataset_path=DATASET_PATH, dataset_name=DATASET_NAME):
    csv_path = os.path.join(dataset_path, dataset_name)
    return pd.read_csv(csv_path)

test_dataset = load_data()
test_dataset.head()
```

```python
DATASET_PATH = "datasets/Titanic"
DATASET_NAME = "train.csv"
DATASET_URL = DATASET_PATH + DATASET_NAME

def fetch_data(dataset_url=DATASET_URL, dataset_path=DATASET_PATH):
    if not os.path.isdir(dataset_path):
        os.makedirs(dataset_path)

train_dataset = fetch_data()

def load_data(dataset_path=DATASET_PATH, dataset_name=DATASET_NAME):
    csv_path = os.path.join(dataset_path, dataset_name)
    return pd.read_csv(csv_path)

train_dataset = load_data()
train_dataset.head()
```

```python
# View the Data
import matplotlib.pyplot as plt
train_dataset.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
train_dataset.describe()
```

```python
#Break out survival rates based on class & sex
fig = plt.figure(figsize=(18,4), dpi=1600)
alpha_level = 0.65

ax1=fig.add_subplot(141)
female_highclass = train_dataset.Survived[train_dataset.Sex == 'female'][train_dataset.Pclass != 3].value_counts()
female_highclass.plot(kind='bar', label='female, highclass', color='#FA2479', alpha=alpha_level)
ax1.set_xticklabels(["Survived", "Died"], rotation=0)
ax1.set_xlim(-1, len(female_highclass))
plt.title("Who Survived? with respect to Gender and Class"); plt.legend(loc='best')

ax2=fig.add_subplot(142, sharey=ax1)
female_lowclass = train_dataset.Survived[train_dataset.Sex == 'female'][train_dataset.Pclass == 3].value_counts()
female_lowclass.plot(kind='bar', label='female, low class', color='pink', alpha=alpha_level)
ax2.set_xticklabels(["Died","Survived"], rotation=0)
ax2.set_xlim(-1, len(female_lowclass))
plt.legend(loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
male_lowclass = train_dataset.Survived[train_dataset.Sex == 'male'][train_dataset.Pclass == 3].value_counts()
male_lowclass.plot(kind='bar', label='male, low class',color='lightblue', alpha=alpha_level)
ax3.set_xticklabels(["Died","Survived"], rotation=0)
ax3.set_xlim(-1, len(male_lowclass))
plt.legend(loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
male_highclass = train_dataset.Survived[train_dataset.Sex == 'male'][train_dataset.Pclass != 3].value_counts()
male_highclass.plot(kind='bar', label='male, highclass', alpha=alpha_level, color='steelblue')
ax4.set_xticklabels(["Died","Survived"], rotation=0)
ax4.set_xlim(-1, len(male_highclass))
plt.legend(loc='best')

plt.show()
```

```python
# Get rid of name, ticket number, and cabin
X_train = train_dataset.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
X_train = X_train.iloc[:, :].values
y_train = train_dataset.iloc[:, 1].values
X_test = test_dataset.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
X_test = X_test.iloc[:, :].values

print(X_test[:, 2])
```

```python
# Taking care of missing data in age column
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(X_train[:, 2].reshape(-1, 1))
X_train[:, 2] = imputer.transform(X_train[:, 2].reshape(-1, 1)).reshape(-1)

imputer = imputer.fit(X_test[:, 2].reshape(-1, 1))
X_test[:, 2] = imputer.transform(X_test[:, 2].reshape(-1, 1)).reshape(-1)

X_test[152, 5] = 0
```

```python
# Encoding categorical data
# Encoding the gender and embarking
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import math

# Encode gender & embark
labelencoder_tr = LabelEncoder()
X_train[:, 1] = labelencoder_tr.fit_transform(X_train[:, 1])
onehotencoder_ts = OneHotEncoder(categorical_features = [1])
X_train = onehotencoder_ts.fit_transform(X_train).toarray()

labelencoder_ts = LabelEncoder()
X_test[:, 1] = labelencoder_ts.fit_transform(X_test[:, 1])
onehotencoder_tr = OneHotEncoder(categorical_features = [1])
X_test = onehotencoder_tr.fit_transform(X_test).toarray()
```

```python
# Scale features and Create Kernel SVM Classifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

#sc_X = StandardScaler()
#X_train_sc = sc_X.fit_transform(X_train)
#X_test_sc = sc_X.transform(X_test)

svm_clf = SVC(kernel='rbf', random_state=0)
svm_clf.fit(X_train, y_train)

print(X_train.shape)
print(X_test.shape)

y_pred_svm = svm_clf.predict(X_test)
```

```python
# Check Performance
acc_svm_clf = round(svm_clf.score(X_train, y_train) * 100, 2)
print(acc_svm_clf)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(n_estimators=100, random_state=0)
forest_clf.fit(X_train, y_train)
y_pred_rf = forest_clf.predict(X_test)

acc_forest_clf = round(forest_clf.score(X_train, y_train) * 100, 2)
print(acc_forest_clf)
```
